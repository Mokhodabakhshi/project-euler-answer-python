```python
import math
```

```python
a = 20
```

```python
def is_prime(x):
    for i in range(2,int(math.sqrt(x))+1):
        if x % i == 0:
            return False
    return True
```

```python
prime = []
for i in range(2,a+1):
    if is_prime(i):
        prime.append(i)
prime = [2,3]
```

```python
res = []
for val in prime:
    p = 1
    temp = val
    while True:
        temp *= val
        if temp > a:
            print(f"{val}^{p}")
            res.append(temp)
            break
        p += 1
   
```

```python
for i in range(len(prime)):
    res[i] = res[i]//prime
    
```
